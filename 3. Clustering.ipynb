```python
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine('postgresql+psycopg2://postgres:0000@localhost:5434/Marketing_Campaign_Project')

query = "SELECT * FROM marketing_campaign_preprocessed_nostd"

with engine.begin() as connection:
    df_nostd = pd.read_sql(query, connection)

df_nostd
```

```python
selected_features = ['Education', 'Marital_Status', 'Income', 'Recency', 'MntCoke', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts','MntGoldProds', 
                     'NumDealsPurchases', 'NumWebPurchases','NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth','AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1',
                     'AcceptedCmp2', 'Complain', 'Response', 'DayofWeek', 'Day', 'Month', 'Year', 'Age', 'Age_Group', 'N_Kids', 'Total_Response', 'Total_Visits','Conversion_Rate', 
                     'Total_Spent', 'Total_Campaign','Membership_Age']
```

```python
numerical = ['Income', 'Recency', 'MntCoke', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts','MntGoldProds', 'NumDealsPurchases', 'NumWebPurchases',
             'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth','DayofWeek', 'Day', 'Month', 'Year', 'Age', 'N_Kids', 'Total_Response', 'Total_Visits',
             'Conversion_Rate', 'Total_Spent', 'Total_Campaign']
categorical = ['Education', 'Marital_Status','AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1', 'AcceptedCmp2', 'Complain', 'Response', 'Age_Group']
```

```python
def get_min_value(col):
    if pd.api.types.is_numeric_dtype(col) or pd.api.types.is_datetime64_any_dtype(col):
        return col.min()
    return "-"

def get_max_value(col):
    if pd.api.types.is_numeric_dtype(col) or pd.api.types.is_datetime64_any_dtype(col):
        return col.max()
    return "-"

def get_mean_value(col):
    if pd.api.types.is_numeric_dtype(col):
        return col.mean()
    return "-"

def get_median_value(col):
    if pd.api.types.is_numeric_dtype(col):
        return col.median()
    return "-"

data_summary = pd.DataFrame({
    'Total Null': df.isna().sum(),
    'Percentage Null': df.isna().sum() * 100 / len(df),
    'Number of Unique Values': df.nunique(),
    'Data Type': df.dtypes,
    'Unique Values': df.apply(lambda col: col.dropna().unique().tolist()),
    'Mode': df.apply(lambda col: col.mode().iloc[0] if not col.mode().empty else None),
    'Min': df.apply(get_min_value),
    'Max': df.apply(get_max_value),
    'Mean': df.apply(get_mean_value),
    'Median': df.apply(get_median_value)
})


# Format Percentage Null
data_summary['Percentage Null'] = data_summary['Percentage Null'].apply(
    lambda x: f"{x:.2f}%" if x != 0 else "0%"
)

data_summary
```

# Feature Selection

```python
from sklearn.feature_selection import mutual_info_classif
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5, random_state=42)  # Sesuaikan jumlah cluster
df['Cluster'] = kmeans.fit_predict(df[selected_features])

# Hitung Mutual Information terhadap hasil clustering
mi_scores = mutual_info_classif(df[numerical + categorical], df['Cluster'], discrete_features='auto')

# Simpan skor Mutual Information
mi_scores = pd.Series(mi_scores, index=numerical + categorical).sort_values(ascending=False)
```

```python
mi_scores
```

Let's take 75% of features (percentiles)

```python
threshold = mi_scores.quantile(0.75)
selected_features = mi_scores[mi_scores > threshold].index.tolist()
```

```python
selected_features
```

# K-Means and Elbow Method of Best Features

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score

inertias = []
silhouette_scores = []

cluster_range = range(2, 15)

for i in cluster_range:
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(df[selected_features])
    labels = kmeans.fit_predict(df[selected_features])
    inertias.append(kmeans.inertia_)

    silhouette_scores.append(silhouette_score(df[selected_features], labels))

fig, ax1 = plt.subplots(figsize=(8,5))
ax1.plot(cluster_range, inertias, marker='o', label="Inertia", color='b')
ax1.set_xlabel("Number of clusters")
ax1.set_ylabel("Inertia", color='b')
ax1.tick_params(axis='y', labelcolor='b')

ax2 = ax1.twinx()
ax2.plot(cluster_range, silhouette_scores, marker='s', linestyle='dashed', label="Silhouette Score", color='orangered')
ax2.set_ylabel("Silhouette Score", color='orangered')
ax2.tick_params(axis='y', labelcolor='orangered')
```

```python
print('Shilouette score at 9 cluster',silhouette_scores[7])
```

We can take 9 as the best number of cluster since the position is in elbow and the shillouette score is starting to decrease at that point.

# Use best cluster, save to dataframe

```python
optimal_k = 9 

kmeans = KMeans(n_clusters=optimal_k, random_state=42)
pred = kmeans.fit_predict(df[selected_features])
df['Cluster'] = pred
df_nostd['Cluster'] = pred
```

# K-Means and Elbow Method according to plan
The plan is to group customer based on income, total spending, and age. Let's se if the clustering result is better.

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score

inertias2 = []
silhouette_scores2 = []

cluster_range = range(2, 11)

for i in cluster_range:
    kmeans2 = KMeans(n_clusters=i)
    kmeans2.fit(df[['Income', 'Total_Spent','Age']])
    labels2 = kmeans2.fit_predict(df[['Income', 'Total_Spent','Age']])
    inertias2.append(kmeans2.inertia_)

    silhouette_scores2.append(silhouette_score(df[['Income', 'Total_Spent','Age']], labels2))

fig, ax1 = plt.subplots(figsize=(8,5))
ax1.plot(cluster_range, inertias2, marker='o', label="Inertia", color='b')
ax1.set_xlabel("Number of clusters")
ax1.set_ylabel("Inertia", color='b')
ax1.tick_params(axis='y', labelcolor='b')

ax2 = ax1.twinx()
ax2.plot(cluster_range, silhouette_scores2, marker='s', linestyle='dashed', label="Silhouette Score", color='orangered')
ax2.set_ylabel("Silhouette Score", color='orangered')
ax2.tick_params(axis='y', labelcolor='orangered')
```

It is clear that the clustering with best feature perform better since the intertia and shilouette score are more stable.
